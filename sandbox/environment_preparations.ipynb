```python
# ==============================================================================
# RESET do projeto para limpar dados antigos
# ==============================================================================
print("🔥 Iniciando destruição e reconstrução do catálogo...")

# 1. Deleta o catálogo inteiro e tudo o que estiver dentro dele (tabelas, schemas, volumes)
spark.sql("DROP CATALOG IF EXISTS ecommerce_intelligence CASCADE")
print(" ✅ Catálogo antigo removido com sucesso (CASCADE).")

# 2. Recria o Catálogo do zero no Unity Catalog
spark.sql("CREATE CATALOG ecommerce_intelligence")
print(" ✅ Novo catálogo 'ecommerce_intelligence' criado.")

# 3. Cria o Schema de Ingestão (Bronze) e o Schema de Landing
spark.sql("CREATE SCHEMA ecommerce_intelligence.landing")
spark.sql("CREATE SCHEMA ecommerce_intelligence.bronze")
spark.sql("CREATE SCHEMA ecommerce_intelligence.silver")
spark.sql("CREATE SCHEMA ecommerce_intelligence.gold")
print(" ✅ Schemas criados.")

# 4. Cria os Volumes necessários para os nossos arquivos e checkpoints
spark.sql("CREATE VOLUME ecommerce_intelligence.landing.landing_files")
spark.sql("CREATE VOLUME ecommerce_intelligence.bronze._checkpoints")
print(" ✅ Volumes de Landing Files e Checkpoints recriados limpos.")

print("\n✨ Ambiente 100% resetado e pronto!")
```
